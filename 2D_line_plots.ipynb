# 2D Line Plots

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x_start, x_end = 1, -1
n = 20
x = np.linspace(x_start, x_end, n)

# equation of a circle
y = np.sqrt(1 - x**2)

# create figure and grab fig, ax objects
fig, ax = plt.subplots(1,2, figsize=[12,6])
fig.suptitle("$\pi$ = {}".format(np.pi), fontsize=22)
fig.subplots_adjust(wspace=0.5) # adjust width between plots

# plot semi-circle
ax[0].set_title("semi-circle")
ax[0].plot(x, y)
ax[0].axis("equal")
ax[0].set_xlabel("x", fontsize=18)
ax[0].set_ylabel("$y=\sqrt{1-x^2}$", fontsize=18)

# plot full circle
ax[1].set_title("circle")
ax[1].axis("equal")
ax[1].plot(x, y, label="top")
ax[1].plot(x, -y, label="bot")
ax[1].set_xlabel("x", fontsize=18)
ax[1].set_ylabel("y = $\pm \sqrt{1-x^2}$", fontsize=18)
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# sin and cosine graphs
theta_start, theta_end = -np.pi, np.pi
n = 40
theta = np.linspace(theta_start, theta_end, n)

y1 = np.sin(theta)
y2 = np.cos(theta)

# create fig, adjust size and width
fig, ax = plt.subplots(2, 2, figsize=[12,12])
fig.subplots_adjust(hspace=0.5, wspace=0.5)
fig.delaxes(ax[1,1]) # remove graph in 2nd row 2nd col
fig.suptitle("Sin and Cos", fontsize=22)

# plot the graphs
ax[0, 0].plot(theta, y1, linestyle="--", marker="o")
ax[0, 0].set_ylabel("$sin(\\theta)$", fontsize=18)
ax[0, 0].set_xlabel("$\\theta$", fontsize=18)

ax[0, 1].plot(theta, y2)
ax[0, 1].set_ylabel("$Cos(\\theta)$", fontsize=18)
ax[0, 1].set_xlabel("$\\theta$", fontsize=18)

ax[1, 0].plot(theta, y1, label="Sin")
ax[1, 0].plot(theta, y2, label="Cos")
ax[1, 0].set_xlabel("$\\theta$", fontsize=18)
ax[1, 0].set_ylabel("y", fontsize=18)
ax[1, 0].legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# generate 1D mesh
n_nodes = 9
x_start = -1
x_end = 1
x_nodes = np.linspace(x_start, x_end, n_nodes)
j = 4 # particular jth node

# create figure and grab fig, axis handles
fig, ax = plt.subplots(figsize=[12,6])

# plot nodes and function
ax.plot([x_nodes[j], x_nodes[j+1]], [-1, 1])
ax.plot(x_nodes, np.zeros(np.size(x_nodes)), marker="o", markersize="8")


# mark jth and (j+1)th node
ax.text(x_nodes[j],-0.1 ,"$x_j$", fontsize=18, horizontalalignment="center")
ax.text(x_nodes[j+1],-0.1, "$x_{j+1}$", fontsize=18, horizontalalignment="center")

# axis limits
ax.set_xlim(x_start, x_end)
ax.set_ylim(-1.1, 1.1)

# axis labels
ax.set_xlabel("$x$", fontsize=18)
ax.set_ylabel("$ \\xi $", fontsize=18)
ax.grid(True)
```
