# 3D Plots

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d

# 3 windings of a spiral
z = np.linspace(0, 6*np.pi, 75)
x = np.cos(z)
y = np.sin(z)

fig = plt.figure(figsize=[13,10])

# add 3d subplot in fig with 1 row 1 col in position index 2
ax1 = fig.add_subplot(1, 1, 1, projection='3d')

ax1.plot(x, y, z, marker='o')
ax1.set_xlabel("$x=cos\\theta$", fontsize=18)
ax1.set_ylabel("$y=sin\\theta$", fontsize=18)
ax1.set_zlabel("$z=\\theta$", fontsize=18)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d

x = np.linspace(-1.5, 1.5, 250)
y = x

X, Y = np.meshgrid(x, y)
Z = -(X**2 + Y**2) + 1

fig = plt.figure(figsize=[18, 13])
fig.suptitle("Downward paraboloid \n $z = -x^2 - y^2 + 1 $", fontsize=20)

ax1 = fig.add_subplot(2,2,1, projection='3d')
surf1 = ax1.plot_surface(X, Y, Z, cmap='YlOrRd')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z');
ax1.grid(False)
fig.colorbar(surf1, shrink=0.7, aspect=10)
ax1.set_title("shrink = 0.7")


ax2 = fig.add_subplot(2,2,2, projection='3d')
surf2 = ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_title("aspect = 20")
fig.colorbar(surf2, shrink=0.5, aspect=20)

ax3 = fig.add_subplot(2, 2, 3, projection='3d')
surf3 = ax3.plot_surface(X, Y, Z, color="blue")
```
